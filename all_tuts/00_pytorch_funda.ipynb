# Start with operations on tensor baby, wohojojoj

```python
# addition 
# addtion will still happen if the dtype is changed
a = torch.tensor([1, 2, 3], dtype=int)
b = torch.tensor([4, 2, 4], dtype=float)
# a + b

aa = torch.tensor([[1, 2, 3], [5, 4, 6]])
print(aa + 10.0)
print(aa * 10)
```

```python
# element wise multiplication : not matrix multiplication
# because if you wanna do mat mul, then n *m and m * n should be there
a = torch.tensor([[2, 3, 4], 
                  [4, 9, 7]])
b = torch.tensor([[3, -1, 5], 
                  [-4, 19, 2]])

a * b
```

```python
torch.mul(a, b)
```

```python
torch.multiply(a, b)
```

```python
torch.matmul(a, b.T)   # this is matrix mul
```

```python
a.T @ b
```

```python
a.matmul(a.T)
```

```python
a @ a.T
```

```python
# we can also use torch.mm short for torch.matmul 
torch.mm(a, b.T)
```
